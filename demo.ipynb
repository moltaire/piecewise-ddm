# Binary case with one process between two absorbing boundaries

```python
thresholds = [-1, 1]
```

```python
Ts = np.array([50, 75, 100, 150])
ds = np.array([-0.01, 0.01, -0.02, 0.02])
σs = np.array([0.05, 0.05, 0.05, 0.05])
```

```python
m = pw.pwddm.PieceWiseDDM(ds=ds, σs=σs, Ts=Ts, thresholds=thresholds)
```

```python
data = m.simulate(repetitions=500)
```

```python
data.plot();
```
